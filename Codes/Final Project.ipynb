# Check point 3: Model Training

We have selected to use Decision Tree Classifier for our baseline model to predict the number of goals. 
### Model 1: Decision Tree Classifier (Base line)
Below is a helper method for training and fitting a Decision Tree Classifier model.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix
```

```python
def decisionTree(xTrain, xTest, yTrain, yTest,firstTime): 
    dTree = DecisionTreeClassifier()
    dTree.fit(xTrain, yTrain)

    yTrainPred = dTree.predict(xTrain)
    yTestPred = dTree.predict(xTest)

    r2Test = dTree.score(xTest, yTest)
    mseTest = mean_squared_error(yTest, yTestPred)
    if(firstTime == 1):
        print("Decision Tree Classifier:")
        printGraphs(dTree,xTrain, xTest, yTrain, yTest, yTrainPred, yTestPred)
    
    return r2Test, mseTest
```

We have selected to use k-Nearest Neighbors and Random Forest for the other 2 models for this prediction.
### Model 2: K-Nearest Neighbors Classifier
K-Nearest Neighbors classificer that uses a majority vote of the nearest neighbors of each point . We have selected to usethe default paramaters for this model. K-Nearest Neighbors helper method.

```python
def kNeighbors(xTrain, xTest, yTrain, yTest,firstTime): 
    knn = KNeighborsClassifier()
    knn.fit(xTrain, yTrain)

    yTrainPred = knn.predict(xTrain)
    yTestPred = knn.predict(xTest)

    r2Test = knn.score(xTest, yTest)
    mseTest = mean_squared_error(yTest, yTestPred)
    if(firstTime == 1):
        print("K-Nearest Neighbors Classifier:")
        printGraphs(knn,xTrain, xTest, yTrain, yTest, yTrainPred, yTestPred)
    
    return r2Test, mseTest
```

### Model 3: Random Forest Classifier
Random Forest uses meta estimator that fits a number of decision tree classifiers and uses averaging to improve accuracy and control over-fitting We have choose to have a max depth of 4 for this version of the model. The value 4 was choosen here so that we do not need to wait for a long time before it is done training and fitting.Random forest model helper method. 

```python
def randomForest(xTrain, xTest, yTrain, yTest,firstTime): 
    forest = RandomForestClassifier(max_depth = 4)
    forest.fit(xTrain, yTrain)

    yTrainPred = forest.predict(xTrain)
    yTestPred = forest.predict(xTest)

    r2Test = forest.score(xTest, yTest)
    mseTest = mean_squared_error(yTest, yTestPred)
    if(firstTime == 1):
        print("Random Forest Classifier:")
        printGraphs(forest,xTrain, xTest, yTrain, yTest, yTrainPred, yTestPred)
    
    return r2Test, mseTest
```

This is helper method to print the graph

```python
def printGraphs(model,xTrain, xTest, yTrain, yTest, yTrainPred, yTestPred):
    print("Goodness of Fit of Model \tTrain Dataset")
    print("Classification Accuracy \t:", model.score(xTrain, yTrain))
    print("Mean Squared Error (MSE) \t:", mean_squared_error(yTrain, yTrainPred))
    print()

    print("Goodness of Fit of Model \tTest Dataset")
    print("Classification Accuracy \t:", model.score(xTest, yTest))
    print("Mean Squared Error (MSE) \t:", mean_squared_error(yTest, yTestPred))
    print()

    f, axes = plt.subplots(1, 2, figsize=(24, 12))
    axes[0].scatter(yTrain, yTrainPred, color = "blue")
    axes[0].plot(yTest, yTest, 'w-', linewidth = 1)
    axes[0].set_xlabel("True values of the Response Variable (Train)")
    axes[0].set_ylabel("Predicted values of the Response Variable (Train)")
    axes[1].scatter(yTest, yTestPred, color = "green")
    axes[1].plot(yTest, yTest, 'w-', linewidth = 1)
    axes[1].set_xlabel("True values of the Response Variable (Test)")
    axes[1].set_ylabel("Predicted values of the Response Variable (Test)")
    
    f, axes2 = plt.subplots(1, 2, figsize=(12, 4))
    axes2[0].set_xlabel("True values of the Response Variable (Train)")
    axes2[0].set_ylabel("Predicted values of the Response Variable (Train)")
    axes2[1].set_xlabel("True values of the Response Variable (Train)")
    axes2[1].set_ylabel("Predicted values of the Response Variable (Train)")
    sb.heatmap(confusion_matrix(yTrain, yTrainPred),
               annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes2[0])
    sb.heatmap(confusion_matrix(yTest, yTestPred), 
               annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes2[1])
    plt.show()
```

### Training and Fitting
Now we will be using the helper model methods we will be train and fitting the models over 100 times to get a more consistent result on the Accuracy Score and MSE value. Each time we will split the data up, train and fit returning the Accuracy Score and MSE values. We will only print the graph once to show how the graph will look like. Printing the graph 100 times is hard to see and understand. All 3 models will use the same dataset.

```python
i = 1
dTreeAScore = 0
dTreeMse = 0
knnAScore = 0
knnMse = 0
forestAScore = 0
forestMse = 0
while(i != 100):
    xTrain, xTest, yTrain, yTest = train_test_split(predictors, response, test_size = 0.20)
    dTreeResult = decisionTree(xTrain, xTest, yTrain, yTest,i)
    dTreeAScore += dTreeResult[0]
    dTreeMse += dTreeResult[1]
    knnResult = kNeighbors(xTrain, xTest, yTrain, yTest,i)
    knnAScore += knnResult[0]
    knnMse += knnResult[1]
    forestResult = randomForest(xTrain, xTest, yTrain, yTest,i)
    forestAScore += forestResult[0]
    forestMse += forestResult[1]
    i+=1
```

```python
print("Average of Linear Regression (over 100 times) Accuracy Score:",dTreeAScore/i," MSE:",dTreeMse/i)
print("Average of k-Nearest-Neighbors (over 100 times) Accuracy Score:",knnAScore/i," MSE:",knnMse/i)
print("Average of Random Forest (over 100 times) Accuracy Score:",forestAScore/i," MSE:",forestMse/i)
```

### Result
From the value above, we can see that random forest have the highest Accuracy Score and the lowest Mean Squared Error. Therefore we will selecting Random Forest for the model and further tuning its hyperparameter 

# Check Mark 4: Hyperparameter Tuning 
We choose to use Grid Search to tune our parameters and the confusion matrix to evaluate the performance together with the use of the Accuracy Score and MSE.

```python
forest = RandomForestClassifier(max_depth = 4)
forest.fit(xTrain, yTrain)

yTrainPred = forest.predict(xTrain)
yTestPred = forest.predict(xTest)

r2Test = forest.score(xTest, yTest)
mseTest = mean_squared_error(yTest, yTestPred)
printGraphs(forest,xTrain, xTest, yTrain, yTest, yTrainPred, yTestPred)
```

### Performance Tuning
We used the grid search to improve the parameter of the random tree classifier. We will be improving the n_estimators, max_depth and the max_leaf_nodes. We first get the n_estimators then we get the max_depth and the max_leaf_nodes

```python
xTrain, xTest, yTrain, yTest = train_test_split(predictors, response, test_size = 0.20)
gridSearch = GridSearchCV(RandomForestClassifier(),
                           param_grid={"n_estimators":range(50,501,10)})
gridSearch.fit(xTrain, yTrain)
print(gridSearch.best_estimator_)
```

#### Graph for n_estimators
Below is the graph to show the different values of n_estimator used. The values given in the tuning starts from 50 to 501 and increment by 10 each time

```python
graphEstiators=sb.lineplot(x=range(50,501,10), y=gridSearch.cv_results_["mean_test_score"])
graphEstiators.set_title(f'Best number of n_estimators at {range(50,501,10)[gridSearch.best_index_]}', size=15);
estimator = range(50,501,10)[gridSearch.best_index_]
```

#### Max Depth and Max Leaf Node
For both the max_depth and the max_leaf_nodes the values we used is starts from 3 to 9. We also use back the value of n_estimator that we found before in this tuning.

```python
gridSearch2 = GridSearchCV(RandomForestClassifier(),
                           param_grid={"n_estimators":[estimator],"max_depth":range(3,10),"max_leaf_nodes":range(3,10)})
gridSearch2.fit(xTrain, yTrain)
print(gridSearch2.best_estimator_)
```

Here we are converting the result from the gridSearch into a a dataframe for graph plotting

```python
i = 3
j = 3
counter = 0;
listColumn = []
while(i<10):
    j=3
    while(j<10):
        column = [None]*3
        column[0] = i
        column[1] = j
        column[2] = gridSearch2.cv_results_["mean_test_score"][counter]
        j+=1
        listColumn.append(column)
        counter+=1
    i+=1
dfResult = pd.DataFrame(listColumn, columns=["max_depth","max_leaf_nodes","score"])
```

### Graph for Max Depth and Max Leaf node
Here we are plotting the graph of to show different result of the parameters

```python
depth = listColumn[gridSearch2.best_index_][0]
maxLeafNode = listColumn[gridSearch2.best_index_][1]

graphDepthLeaf=sb.lineplot(x="max_depth", y="score",hue="max_leaf_nodes", data = dfResult, palette = "tab10")
graphDepthLeaf.set_title(f'Best max depth is {depth} and max leaf node is {maxLeafNode}', size=15);
graphDepthLeaf.legend(loc='upper right')
```

### New Random Forest Classifier with tuned values

```python
def randomForestWithTuned(xTrain, xTest, yTrain, yTest,firstTime): 
    forest = RandomForestClassifier(max_depth = depth, max_leaf_nodes = maxLeafNode, n_estimators = estimator)
    forest.fit(xTrain, yTrain)

    yTrainPred = forest.predict(xTrain)
    yTestPred = forest.predict(xTest)

    r2Test = forest.score(xTest, yTest)
    mseTest = mean_squared_error(yTest, yTestPred)
    if(firstTime == 1):
        print("Random Forest Classifier with tuned values:")
        printGraphs(forest,xTrain, xTest, yTrain, yTest, yTrainPred, yTestPred)
    
    return r2Test, mseTest
```

```python
i = 1
dTreeAScore = 0
dTreeMse = 0
forestTunedAScore = 0
forestTunedMse = 0
while(i != 100):
    xTrain, xTest, yTrain, yTest = train_test_split(predictors, response, test_size = 0.20)
    dTreeResult = decisionTree(xTrain, xTest, yTrain, yTest,i)
    dTreeAScore += dTreeResult[0]
    dTreeMse += dTreeResult[1]
    forestTunedResult = randomForestWithTuned(xTrain, xTest, yTrain, yTest,i)
    forestTunedAScore += forestTunedResult[0]
    forestTunedMse += forestResult[1]
    i+=1
```

```python
concl = pd.DataFrame(merged2[["HomeTeam","AwayTeam","FTHG", "FTAG","FTR","HTHG","HTAG","HTR","HS","AS","HST","AST","HC","AC","HF","AF"]])
concl
```

```python
teams = list(concl.loc['17/18']['HomeTeam'].unique())
teamdata = {team:concl[np.logical_or(concl["HomeTeam"]==team, concl["AwayTeam"]==team)] for team in teams}
```

```python
teams
```

```python
seasons = ['17/18','18/19','19/20','20/21','21/22']
teams = ['Man City', 'Crystal Palace']
teamgoals = {}
avg_goals = {}
avg_shots = {}  # added new dictionary for average shots
for team in teams:
    for season in seasons:
        matches_played = len(teamdata[team].loc[teamdata[team].index.get_level_values(0) == season].loc[(teamdata[team]["HomeTeam"] == team) | (teamdata[team]["AwayTeam"] == team)])
        goals_scored = teamdata[team].loc[teamdata[team].index.get_level_values(0).str.contains(season)].loc[teamdata[team]["HomeTeam"] == team, 'FTHG'].sum() + teamdata[team].loc[teamdata[team].index.get_level_values(0).str.contains(season)].loc[teamdata[team]["AwayTeam"] == team, 'FTAG'].sum()
        teamgoals[season + " " + team] = goals_scored
        shots_on_target = teamdata[team].loc[teamdata[team].index.get_level_values(0) == season][['HST', 'AST']].sum().sum()
        avg_shots[season + " " + team] = shots_on_target / matches_played  # calculate average shots per match
        avg_goals[season + " " + team] = goals_scored / matches_played
```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

# plot the first bar chart
ax1.bar(x=[i-0.2 for i in range(len(avg_shots))], height=avg_shots.values(), width=0.4, label='Average Shots on Target per Match')

# plot the second bar chart
ax2 = ax1.twinx()
ax2.bar(x=[i+0.2 for i in range(len(avg_goals))], height=avg_goals.values(), width=0.4, color='orange', label='Average Goals Scored per Match')

# set the x-axis labels
ax1.set_xticks(range(len(avg_shots)))
ax1.set_xticklabels(avg_shots.keys(), rotation=45, ha='right')

# set the y-axis labels and scales
ax1.set_ylabel('Average Shots on Target per Match')
ax2.set_ylabel('Average Goals Scored per Match')
ax1.set_ylim(0, max(avg_shots.values())*1.2)
ax2.set_ylim(0, max(avg_goals.values())*1.2)

# set the chart title and legend
ax1.set_title('Average Shots on Target and Average Goals Scored by Team by Season')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
```

As seen from the "Shots on Target & Goals Scored"
graph above, we can see how, generally, more shots on target translates to more goals scored. However, better teams such as Manchester City are able to convert these shots on target to actual goals more often. Teams should thus not only aim to maximise shots on target by creating chances, but players should be able to efficiently and consistently convert these opportunities in matches to score.
